## 2. Get dicts of songs from playlist

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
tracks = get_playlist_tracks("spotify","2gcYSwUOBfhYI52ACzlPko")
```

```python
tracks
```

## 3. Get ID's of the songs

```python
songs_id_list = []
for i in tracks:
    songs_id_list.append(i['track']['id'])
```

```python
songs_id_list
```

## 4. Get A.F. of songs

```python
audio_list = []
for i in songs_id_list:
    audio_list.append(sp.audio_features(i)[0])
```

```python
audio_list
```

## 5. Save to DF

```python
audio_list_df = pd.DataFrame(audio_list)
```

```python
audio_list_df
```

```python
audio_list_df.to_csv("audio_list_df_file.csv",index=False)
```
